## Q1 - Does the dataset have any missing data? If the answer is positive, which column have the most missing data?

```python
# Check missing values
missing_values = df.isnull().values.any()
# Column with the most missing values
column = df.isnull().sum().idxmax()
pd.DataFrame({
    'Check missing values':missing_values,
    'Column with the most missing values':column
},index=['0'])
```

## Q2 - Drop 'company' column from DataFrame.

```python
df.drop('company',axis=1,inplace=True)
```

```python
df.columns
```

## Q3 - Which country have the most passengers? Define the top 5 countries.

```python
df['Total passengers'] = (df['adults']+df['children']+df['babies']) - df['is_canceled']
passengers = df.groupby('country')[['adults','children','babies','Total passengers']].sum().nlargest(5,'Total passengers')
passengers
```

## Q7 - Define the name of the passenger who have the maximum average daily rate (ADR). How much is its price?

```python
df.head(2)
```

```python
name = df.loc[df['adr'].idxmax(),'name']
price = df['adr'].max()
pd.DataFrame({
    'name':name,
    'price':price
},index=['ADR'])
```

## Q8 - Average of total ADRs with 2 decimals

```python
df['adr'].mean().round(2)
```

## Q9 - Define the average of number of nights stayed

```python
average_night_stayed = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
average_night_stayed.mean().round(2)
```

## Q10 - Define the name and e-mail of people who had 5 special requests

```python
special_requests = df.loc[df['total_of_special_requests']==5,['name','email','total_of_special_requests']]
special_requests
```

## Q11 - Which first names have the most frequency of last name? Define 5 most frequent family

```python
split_names = df['name'].str.split(' ').str[1] 
split_names.value_counts().nlargest(5)
```

## Q12 - Define the people whom reserved a hotel with most number of babies and children

```python
df['total_babies_children'] = df['babies'] + df['children']
df.loc[df['total_babies_children'].idxmax(),['name','email','total_babies_children']]
```
